```python
# import required libraries
from wwo_hist import retrieve_hist_data
import os
```

Using WWO(World-weather online) API key to make a API call requesting every data from start date to end date to use for training the model

```python
# Chosen 'Mumbai' as the standard region considering the increasing complexity of the problem due to availability of many locations around the country

FREQUENCY = 24
START_DATE = '1-JAN-2021'
END_DATE = '28-JUN-2024'
API_KEY = 'api_key' # Replace with your api_key
LOCATION_LIST = ['mumbai']

hist_weather_data = retrieve_hist_data(API_KEY,
                                LOCATION_LIST,
                                START_DATE,
                                END_DATE,
                                FREQUENCY,
                                location_label = False,
                                export_csv = True,
                                store_df = True)

# the data is saved as a csv file named 'mumbai.csv'
```
